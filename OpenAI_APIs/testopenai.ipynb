```python
from openai import OpenAI
import pandas as pd
from dotenv import load_dotenv
```

```python
import os
OpenAI.api_key=os.getenv("OPENAI_API_KEY")
```

```python
load_dotenv()

client = OpenAI()
models = client.models.list()

df = pd.DataFrame(
    [{"id": model.id, "created": model.created} for model in models.data]
)

df
```

```python
# Chat completion API and Completion API
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "user",
            "content": "What is the capital of India?"
        }
    ]
)

print(response.choices[0].message.content)
```

```python
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "user",
            "content": """Act as a translator. Translate the following text to French: Hello, how are you?"""
        }
    ],
    max_completion_tokens=50,
    temperature=1,
    n=3
)

print(response.choices[0].message.content)
```
